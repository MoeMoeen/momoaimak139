```python
import pandas as pd
import numpy as np
from collections import Counter
from google.colab import files

uploaded = files.upload()


df = pd.read_csv("iris_dataset.csv")
df
```

```python
X = df[["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]].values
y = df["target"].values
```

```python
P_test = X[149]
y_test_true = y[149]

X_train = X[:149]
y_train = y[:149]

P_test, y_test_true
```

```python
def dist(a, b):
    return np.sqrt(np.sum((a - b)**2))
```

```python
distances = np.array([dist(P_test, p) for p in X_train])
```

```python
k = 3
idx = distances.argsort()[:k]
neighbors = y_train[idx]
neighbors
```

```python
pred_k3_raw = Counter(neighbors).most_common(1)[0][0]
pred_k3_raw
```

```python
mu = X_train.mean(axis=0)
sigma = X_train.std(axis=0)

print("mu is:", mu)
print("sigma is:", sigma)
```

```python
X_train_norm = (X_train - mu) / sigma
```

```python
df_temp_raw = pd.DataFrame(X_train)
df_temp_norm = pd.DataFrame(X_train_norm)
df_temp = pd.concat([df_temp_raw, df_temp_norm], axis=1)
df_temp
```

```python
P_test_norm = (P_test - mu) / sigma
```

```python
distances_norm = np.array([dist(P_test_norm, p) for p in X_train_norm])
```

```python
idx_norm = distances_norm.argsort()[:3]
neighbors_norm = y_train[idx_norm]
neighbors_norm
```

```python
pred_k3_norm = Counter(neighbors_norm).most_common(1)[0][0]
pred_k3_norm
```

```python
print("True class:", y_test_true)
print("Predicted (no normalization):", pred_k3_raw)
print("Predicted (Z-score normalization):", pred_k3_norm)
```
