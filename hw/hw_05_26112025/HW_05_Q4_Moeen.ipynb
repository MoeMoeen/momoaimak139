```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
```

```python
iris_dataset_uploaded = files.upload()
```

```python
df = pd.read_csv("iris_dataset.csv")
df
```

```python
mean_table = df.groupby("target").mean()
mean_table
```

```python
std_table = df.groupby("target").std()
std_table
```

```python
range_table = df.groupby("target").max() - df.groupby("target").min()
range_table
```

```python
corr_class0 = df[df["target"]==0].iloc[:, :4].corr()
corr_class0
```

```python
corr_class1 = df[df["target"]==1].iloc[:, :4].corr()
corr_class1
```

```python
corr_class2 = df[df["target"]==2].iloc[:, :4].corr()
corr_class2
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1,3,1)
sns.heatmap(corr_class0, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("0")

plt.subplot(1,3,2)
sns.heatmap(corr_class1, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("1")

plt.subplot(1,3,3)
sns.heatmap(corr_class2, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("2")

plt.tight_layout()
plt.show()
```
