```python
P_test = x[149]
y_test_true = y[149]
```

```python
P_test
```

```python
y_test_true
```

```python
x_train = x[:149]
y_train = y[:149]
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))
```

```python
distances = np.array([euclidean_distance(P_test, point) for point in x_train])
distances
```

```python
index_min = distances.argmin()
index_min
print(distances[index_min])
```

```python
nearest_point = x_train[index_min]
nearest_point
```

```python
idx_min = distances.argmin()
nearest_point = x_train[idx_min]
predicted_class_k1 = y_train[idx_min]
```

```python
print("Test Point =", P_test)
print("Nearest Neighbor =", nearest_point)
print("Predicted Class (k=1) =", predicted_class_k1)
print("True Class =", y_test_true)
```

```python
k = 3
idx_sorted = distances.argsort()[:k]
idx_sorted
```

```python
distances[idx_sorted]
```

```python
nearest_classes = y_train[idx_sorted]
nearest_classes
```

```python
from collections import Counter

print(Counter(nearest_classes))
```

```python
predicted_class_k3 = Counter(nearest_classes).most_common(1)[0][0]
predicted_class_k3
```

```python
print("3 nearest neighbors classes:", nearest_classes)
print("Predicted class (k=3):", predicted_class_k3)
print("True class:", y_test_true)
```
