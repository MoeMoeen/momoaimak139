```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Section 1**

```python
cities = ["Tehran", "Mashhad", "Isfahan", "Tabriz", "Shiraz", "Ahvaz"]
n_cities = len(cities)
n_days = 30
n_rows = n_cities * n_days
```

```python
np.random.seed(42)

temperatures = np.random.uniform(15, 40, size=n_rows)

humidities = np.random.uniform(20, 80, size=n_rows)

rainfalls = np.random.uniform(0, 50, size=n_rows)

```

**Section 2**


```python
city_column = np.repeat(cities, n_days)

day_column = np.tile(np.arange(1, n_days+1), n_cities)

data = {
    "City": city_column,
    "Day": day_column,
    "Temperature": temperatures,
    "Humidity": humidities,
    "Rainfall": rainfalls
}

df = pd.DataFrame(data)

df.head(50)
```

**Section 3**

```python
city_means = df.groupby("City")[["Temperature", "Humidity", "Rainfall"]].mean()
print("میانگین دما، رطوبت و بارندگی هر شهر:")
print(city_means)
```

```python
warmest_city = city_means["Temperature"].idxmax()
coldest_city = city_means["Temperature"].idxmin()

print("گرم‌ترین شهر از نظر میانگین دما:", warmest_city)
print("سردترین شهر از نظر میانگین دما:", coldest_city)
```

```python
rainy_days = df[df["Rainfall"] > 10]
rainy_count_per_city = rainy_days.groupby("City")["Day"].count()

print("Total rainy days across all cities: ")
print(rainy_days.shape[0])
print("="*40)

print("تعداد روزهای با بارندگی بیشتر از 10 میلی‌متر برای هر شهر:")
print(rainy_count_per_city)
```

**Section 4:**

```python
isfahan_data = df[df["City"] == "Isfahan"]
isfahan_data.head()
```

```python
plt.figure(figsize=(6, 4))

plt.scatter(isfahan_data["Temperature"], isfahan_data["Humidity"])

plt.title("Temperature - Humidity Relationship for Esphahan)")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

plt.grid(True)
plt.show()
```

```python
corr_isfahan = isfahan_data[["Temperature", "Humidity"]].corr()
print(corr_isfahan)
```
