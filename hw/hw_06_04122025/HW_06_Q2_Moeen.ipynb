```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
```

```python
breast = load_breast_cancer(as_frame=True)
df = breast.frame.copy() # includes indices + features + target
df
```

```python
print("Columns:\n", df.columns.tolist())
print("\nTarget names:", breast.target_names)
```

```python
X = df.drop(columns=["target"])
y = df["target"]
print("\nShape of X (samples, features):", X.shape)
print("\nType of x:", type(X))
print("\nShape of y:", y.shape)
print("\nType of y:", type(y))
y
```

```python
print("\nFirst 5 rows of features:")
print(X.head())
```

```python
print("\nStatistical summary of features:")
print(X.describe().T)
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances
```

```python
benign_idx = y[y == 1].index[:3]
malignant_idx = y[y == 0].index[:3]

chosen_idx = list(benign_idx) + list(malignant_idx)

X_small = X.loc[chosen_idx].reset_index(drop=True)
y_small = y.loc[chosen_idx].reset_index(drop=True)

print("Chosen sample labels (1=benign, 0=malignant):")
print(y_small.tolist())

```

```python
from sklearn.metrics import pairwise_distances
from pprint import pprint


metrics = ["euclidean", "manhattan", "chebyshev", "cosine"]
distance_results = {}

for m in metrics:
    distance_results[m] = pairwise_distances(X_small, X_small, metric=m)

# inverse covariance of the 6-sample dataset
cov = np.cov(X_small.values, rowvar=False)
VI = np.linalg.inv(cov)

distance_results["mahalanobis"] = pairwise_distances(
    X_small, X_small, metric="mahalanobis", VI=VI
)

print("Distances:\n")
pprint(distance_results)
```

```python
for m, mat in distance_results.items():
    print(f"\n=== {m.upper()} DISTANCE MATRIX ===")
    print(pd.DataFrame(mat))
```
