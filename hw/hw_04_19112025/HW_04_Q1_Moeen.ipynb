```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"] = iris.target

df
```

```python
species_map = {0: "setosa", 1: "versicolor", 2: "virginica"}
df["species_name"] = df["species"].map(species_map)

df
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df["petal_ratio"] = df["petal length (cm)"] / df["petal width (cm)"]
df
```

```python
corr = df.iloc[:, :4].corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Between Iris Features")
plt.show()
```

```python
selected_cols = [
    "sepal length (cm)",
    "sepal width (cm)",
    "petal length (cm)",
    "petal width (cm)",
    "petal_ratio"
]

sns.pairplot(df[selected_cols + ["species_name"]], hue="species_name", corner=True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="species_name", y="petal_ratio")
plt.title("Distribution of New Feature (petal_ratio) by Species")
plt.xlabel("Species")
plt.ylabel("Petal Ratio")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species_name",
    s=80
)
plt.title("Petal Length vs Width by Species")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal_ratio",
    hue="species_name",
    s=80
)
plt.title("Petal Ratio vs Petal Length by Species")
plt.xlabel("Petal Length - cm")
plt.ylabel("Petal Ratio")
plt.show()
```

```python
df_petal_ratio_means = df.groupby("species_name")["petal_ratio"].mean()
df_petal_ratio_means = df_petal_ratio_means.rename('petal_ratio_mean')
df_petal_ratio_means
```
